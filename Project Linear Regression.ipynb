```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("pharmacy_with_pop.csv")
df.head()
```

```python
AgeGroup = '55-59'
MedicationType = 'Psychoanaleptics'
Frequency= 'M'
```

```python
is_Med = df['Medication_Type'] == MedicationType
df2 = df[is_Med]
df2.head()
```

```python
is_Age = df2['Age Group'] == AgeGroup
df2 = df2[is_Age]
df2.head()
```

```python
df2['Delivery Date'] = df['Delivery Date'].astype('datetime64[ns]')
df2.head()
```

```python
res = df2.set_index('Delivery Date').groupby([pd.Grouper(freq=Frequency), 'Total Population'])['Units'].sum().reset_index()

print(res)
```

```python
res['Month'] = 0
res['Month'] = pd.DatetimeIndex(res['Delivery Date']).month
res.head()
```

```python
res['Total Population'] = res['Total Population'].astype(int)
res.dtypes
res = res.reset_index()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X = res[['Total Population', 'Month']].values
y = res['Units'].values
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.3, random_state=0
)
slr = LinearRegression()
slr.fit(X_train, y_train)
y_train_pred = slr.predict(X_train)
y_test_pred = slr.predict(X_test)
print('Slope: %.3f' % slr.coef_[0])
print('Intercept: %.3f' % slr.intercept_)
```

```python
plt.scatter(y_train_pred, y_train_pred - y_train,
    c='blue', marker='o', label='Training data'
)
plt.scatter(y_test_pred,  y_test_pred - y_test,
    c='lightgreen', marker='s', label='Test data'
)
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-0, xmax=1, lw=2, color='red')
plt.show()
```

```python
from sklearn.metrics import median_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print("Mean Squared Error: ",mean_squared_error(y_test, y_test_pred))
errors = abs(y_test_pred-y_test)
print('Mean Absolute Error:', round(np.mean(errors), 2))
print('R2 score: ',r2_score(y_test, y_test_pred))
print('Median Absolute Error: ',median_absolute_error(y_test, y_test_pred))
print('Accuracy: ', slr.score(X_test,y_test)*100)
```
