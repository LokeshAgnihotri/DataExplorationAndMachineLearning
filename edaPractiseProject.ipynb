```python
import numpy as np
import pandas as pd
import sklearn as sk
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
df_pharmacy = pd.read_csv("C:/Users/Acer/Downloads/ds.csv")
```

```python
df_pharmacy.columns = ['Delivery Date','Delivery Time','Pharmacy Number','Pharmacy Postcode','DOB','Gender','CNK','Product name','ATC','Units','Price','Contribution']
df_pharmacy.head()
```

```python
# chechking number of rows and columns
df_pharmacy.shape
```

```python
# cleaning the data
df_pharmacy.isnull().sum()
```

```python
correlation = df_pharmacy.corr()
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns,annot=True)
```

```python
sns.distplot(df_pharmacy['Units'],bins=5)
```

```python
df_pharmacy[['Price']].plot(kind='hist',bins=[0,30,60,90,200,300,400],rwidth=0.8)
plt.show()
```

```python
## comparing year of birth of customer with the price
plt.scatter(df_pharmacy['DOB'],df_pharmacy['Price'])
plt.xlabel('Year of birth')
plt.ylabel('Price')
plt.show()
```

```python
## creating a new column extracting only the delivery year from the date
df_pharmacy['Delivery Year'] = pd.DatetimeIndex(df_pharmacy['Delivery Date']).year
df_pharmacy.head()
```

```python
## comparing year of units with the price
plt.scatter(df_pharmacy['Units'],df_pharmacy['Price'])
plt.xlabel('Units')
plt.ylabel('Price')
plt.show()
```
